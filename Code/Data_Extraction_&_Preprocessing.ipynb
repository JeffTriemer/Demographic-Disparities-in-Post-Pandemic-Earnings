For industry, that's where it gets messy. The options in the questionnaire are as follow

Yet the number of unique values are as follow

```r
length(unique(cleaned_data$industry))
```

First, we title-case the values, as well as replace occurrences of "and" or "And" with "&". We'll then perform some regexp matching to extract

```r
# We convert to title to ensure more consistent, as well as remove trailing spaces
cleaned_data$industry_cleaned <- trimws(tolower(cleaned_data$industry))
# We
industry_mapping <- c(
  '.*education.*' = 'Education (Higher Education)',
  '.*computing.*|.*tech.*' = 'Computing or Tech',
  '.*accounting.*|.*banking.*|.*finance.*' = 'Accounting, Banking & Finance',
  '.*nonprofits.*|.*philanthropy.*' = 'Nonprofits',
  '.*publishing.*|.*media.*|.*digital.*' = 'Media & Digital',
  '.*primary/secondary.*|.*education services.*|.*ed tech.*' = 'Education (Primary/Secondary)',
  '.*law.*|.*legal.*' = 'Law',
  '.*health care.*|.*medical.*' = 'Health care',
  '.*business.*|.*consulting.*' = 'Business or Consulting',
  '.*art.*|.*design.*' = 'Art & Design',
  '.*government.*|.*public administration.*' = 'Government and Public Administration',
  '.*public library.*|.*libraries.*' = 'Libraries',
  '.*retail.*' = 'Retail',
  '.*marketing.*|.*advertising.*|.*pr.*' = 'Marketing, Advertising & PR',
  '.*property.*|.*construction.*' = 'Property or Construction',
  '.*biotech.*|.*pharma.*' = 'Biotech/Pharmaceuticals',
  '.*aerospace.*' = 'Engineering or Manufacturing',
  '.*entertainment.*' = 'Entertainment',
  '.*insurance.*' = 'Insurance',
  '.*hospitality.*' = 'Hospitality & Events',
  '.*transport.*|.*logistics.*' = 'Transport or Logistics',
  '.*medical devices.*' = 'Health care',
  '.*social work.*' = 'Social Work',
  '.*surveying.*' = 'Engineering or Manufacturing',
  '.*phd.*' = 'Education (Higher Education)'
)

library(stringr)

for (pattern in names(industry_mapping)) {
  replacement <- industry_mapping[pattern]
  cleaned_data$industry_cleaned <- str_replace_all(regex(cleaned_data$industry_cleaned, ignore_case=TRUE), pattern, replacement)
}
```

This mapping is quite efficient. We can work more on it, but for now 50% cleaned is good; the rest we'll map to "Other"

```r
length(unique(cleaned_data$industry_cleaned))
```

```r
cleaned_data$industry_cleaned = ifelse(cleaned_data$industry_cleaned %in% industries, cleaned_data$industry_cleaned, "Other")
```

```r
# drop the column "industry" and replace the name "industry_cleaned" by "industry"
if ("industry_cleaned" %in% colnames(cleaned_data)) {
  cleaned_data <- subset(cleaned_data, select = -industry_cleaned)
  colnames(cleaned_data)[colnames(cleaned_data) == "industry_cleaned"] <- "industry"
}
# now we can cast the type industry to factor
cleaned_data$industry <- as.factor(cleaned_data$industry)
head(cleaned_data)
```

## Handling Missing Values

We see below that the most important variable for us, salary, doesn't have any missing values. However, annual_bonus, race, and state do have, and we'll treat them.

```r
processed_data[processed_data == ""] <- NA
# count number of missing values per columns
missing_counts <- colSums(is.na(processed_data))
print(missing_counts)
```

```r
processed_data <- processed_data[!is.na(processed_data$education_level), ]
processed_data <- processed_data[!is.na(processed_data$overall_experience_years), ]
processed_data <- processed_data[!is.na(processed_data$field_experience_years), ]
processed_data <- processed_data[!is.na(processed_data$age), ]
```

Various treatment options for missing values, whether removing them or imputing them (i.e. MICE, mean/median/mode, neighbour techniques etc.) but we'll keep it simple. We can fill empty values for annual_bonus with 0s, drop rows with empty "state" (since it is important for us to know which state someone is from in our study), and for race, replace with "Other".

```r
processed_data$annual_bonus[is.na(processed_data$annual_bonus)] <- 0
processed_data <- processed_data[!is.na(processed_data$state), ]
processed_data$race[is.na(processed_data$race)] <- "Another option not listed here or prefer not to answer"
processed_data <- processed_data[processed_data$race != "Another option not listed here or prefer not to answer", ]
processed_data$race <- as.factor(processed_data$race)
```

We can confirm below no more missing values:

```r
# count number of missing values per columns
missing_counts <- colSums(is.na(processed_data))
print(missing_counts)
```
